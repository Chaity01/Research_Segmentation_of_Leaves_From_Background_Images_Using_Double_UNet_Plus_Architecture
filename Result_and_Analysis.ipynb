# New Section

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Growth of Dice Coefficient Over Time for Each Model
plt.figure(figsize=(10,12))
for filename in csv_files:
  df = pd.read_csv(f"{root_dir}{filename}")
  model_name = filename.split("_")[0]
  #epoch 40
  if df.shape[0] == 40:
    plt.plot(df.index, df['dice_coef'], label=f'{model_name}')


plt.grid(True)
plt.legend(loc='lower right', fontsize='small', title='Model names')
plt.xlabel('Epoch')  # Assuming the index represents epochs
plt.ylabel('Dice Coefficient')
plt.title('Growth of Dice Coefficient Over Time for Each Model')
plt.show()
```

## Test results

```python
import os
test_value_root = "/content/drive/MyDrive/Thesis Rumex Leave-SUB/RumexLeaves/iNaturalist/test_values/"
test_value_files = os.listdir(test_value_root)
print(test_value_files)
model_names = [filename.split("_")[0] for filename in test_value_files]
model_names
```

```python
csv_file = pd.read_csv(f"{test_value_root}{test_value_files[0]}")
metric_names = csv_file.iloc[:, 0]

#Create a table
combined_data = pd.DataFrame()

#Combine Model name
for filename in test_value_files:
  df = pd.read_csv(f"{test_value_root}{filename}", skiprows=1)
  model_name = filename.split("_")[0]
  combined_data[model_name] = df.iloc[:,1].values


combined_data.index = df.iloc[:,0]
combined_data.plot(kind='bar', figsize=(10,6))
plt.title('Comparison of Metrics Across Models')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.xticks(rotation=45)
plt.legend(title='Model')
plt.tight_layout()

# Show the plot
plt.show()
print(combined_data)
```

```python
dice_coef_df = combined_data.loc['Dice Coefficient']
dice_coef_df.plot(kind='bar', figsize=(10, 4))

for index, value in enumerate(dice_coef_df):
    plt.text(index, value, str(round(value, 4)), ha='center', va='bottom')

plt.title('Dice Coefficient Comparison Across Models')
plt.ylabel('Dice Coefficient')
plt.xlabel('Model names')
plt.xticks(rotation=45)
plt.ylim(0.8, max(dice_coef_df) * 1.1)
plt.tight_layout()

# Show the plot
plt.show()
dice_coef_df
```

```python
def bar_by_metric(metric_name):
  metric_df = combined_data.loc[metric_name].sort_values(ascending=False)
  metric_df.plot(kind='bar', figsize=(10, 4))

  for index, value in enumerate(metric_df):
      plt.text(index, value, str(round(value, 4)), ha='center', va='bottom')

  plt.title(f'{metric_name} Comparison Across Models')
  plt.ylabel(metric_name)
  plt.xlabel('Model names')
  plt.xticks(rotation=45)
  plt.ylim(min(metric_df) * 0.7, max(metric_df) * 1.1)
  plt.tight_layout()

  # Show the plot
  plt.show()


bar_by_metric("IoU")
```
