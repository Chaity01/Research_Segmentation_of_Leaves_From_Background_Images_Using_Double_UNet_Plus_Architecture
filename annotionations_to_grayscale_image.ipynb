```python
import xml.etree.ElementTree as ET
import cv2
import numpy as np

def process_annotations( index):
    # Load the annotations file
    tree = ET.parse('/content/drive/MyDrive/VGG16- Rumex Leaves/iNaturalist/annotations.xml')
    root = tree.getroot()[index]

    # Get the image dimensions
    width = int(root.attrib.get('width'))
    height = int(root.attrib.get('height'))
    name = root.attrib.get('name')

    # Create a grayscale image with the same dimensions as the original image
    image = np.zeros((height, width), dtype=np.uint8)

    # Iterate over the polygons in the annotations file
    for polygon in root.iter('polygon'):
        # Get the coordinates of the polygon vertices
        points_str = polygon.attrib.get('points')
        points_splitted = points_str.split(';')
        pnts_tpl = []
        for pnt_str in points_splitted:
            pnt = pnt_str.split(",")
            pnt_x = int(pnt[0].split(".")[0])
            pnt_y = int(pnt[1].split(".")[0])
            pnts_tpl.append((pnt_x, pnt_y))

        # Draw the polygon on the image
        cv2.fillPoly(image, [np.array(pnts_tpl)], (255, 255, 255))

    return image, name

def main():
    #xml_path = 'annotations.xml'
    print("running iteration")
    for index in range(2, 692):
      print('index', index)
      result_image, name = process_annotations( index)
      image_id = name.split(".")[0]
      # Save the image
      cv2.imwrite(f'{root_dir}/masks/{image_id}.png', result_image)

main()
```

```python
import os

masked_image_dir = f"{root_dir}/masks/"
resized_image_dir = f"{root_dir}/masks_resized/"
target_size = (224, 224)

def resize_img(image_name):
  img = cv2.imread(f"{masked_image_dir}{image_name}")
  try:
    resized_img = cv2.resize(img, target_size)
    cv2.imwrite(f"{resized_image_dir}{image_name}", resized_img)
  except Exception as e:
    print("error", e)

for file in os.listdir(masked_image_dir):
  resize_img(file)
  print(file)
```
